```python
import pandas as pd
import numpy as np
main_df = pd.read_csv("survey_responses.csv")
sub_df = pd.read_csv("survey_questions_meta.csv")
```

```python
# Check for possible ID columns
print("Main DF Columns:", main_df.columns.tolist())
print("Sub DF Columns:", sub_df.columns.tolist())
```

```python
df = pd.merge(main_df, sub_df, on="response_id", how="inner")
```

```python
print(df.shape)
df.columns.tolist()
```

```python
df.head()
```

```python
categorical_col = df.select_dtypes(include=['object']).columns.tolist()
# Filter those that start with 'Q'
categorical_cols = [col for col in categorical_col if col.startswith(('Q1', 'Q2', 'Q3', 'Q4', 'Q5'))]
print("Categorical Columns:", categorical_cols)
```

```python
# Apply mapping to all categorical columns
for col in categorical_cols:
  df[col] = df[col].map({"Yes": 1,"No": 0})
  df[col] = df[col].fillna(0).astype(int)
```

```python
df.to_csv('student_survey_cleaned.csv', index=False)
df = pd.read_csv("student_survey_cleaned.csv", encoding="utf-8-sig")
df.head()
```
